```python
import pandas as pd
import numpy as np
from surveyweights import normalize_weights 

survey = pd.read_csv('responses_processed_national_weighted.csv')
survey
```

```python
print('## AGREEMENT, UNWEIGHTED ##')
print(survey['agree_farming_harms_environment'].value_counts(normalize=True) * 100)
print('N = {}'.format(len(survey)))
```

```python
survey['lv_weight_alt'] = survey['lv_weight'] / survey['lv_index']
survey['lv_weight_alt'] = normalize_weights(survey['lv_weight_alt'])
    
print('## AGREEMENT, WEIGH ON GENDER + RACE + AGE + INCOME + REGION + EDUCATION + URBAN/RURAL + 2016 VOTE + SOCIAL TRUST + RELIGION + SPANKING ##')
print(survey['agree_farming_harms_environment'].value_counts(normalize=True) * survey.groupby('agree_farming_harms_environment')['lv_weight'].mean() * 100)
weighted_n = int(np.round(survey['lv_weight_alt'].apply(lambda w: 1 if w > 1 else w).sum()))
print('Weighted N = {}'.format(weighted_n))
```
